```python
def show_info(data, column):
    print(f'Number of unique values for {column}: {data[column].nunique()}\n\n')
    print(f'Unique values for {column}:\n{data[column].unique()}\n\n')
    print(f'Value counts for {column}:\n{data[column].value_counts()}\n\n')
    print(f'Number of NaN-values: {data[column].isna().sum()}')
    
    
def bar(column):
    """
    Plots pie chart and bar chart
    for given categorical feature
    """
    fig, axes = plt.subplots(figsize=(12, 6))
    
    sns.barplot(x=df_1[column].astype(str).value_counts().index[:5],
                y=df_1[column].value_counts(normalize=True).values[:5],
                palette='magma')
    plt.title(f'Top categories for "{column}"',
              size=15)
    plt.ylabel('Proportion')
    
    plt.xticks(rotation=90)
    plt.subplots_adjust(wspace=0.2);
    
    
def dist_box(data, column):
    """
    Plots distribution
    and boxplot for given
    numeric feature
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    sns.kdeplot(data[column], ax=axes[0][0],
                color='r', shade=True)
    axes[0][0].set_title(f'Distribution for "{column}"',
                      size=15)
    sns.boxplot(data[column], ax=axes[0][1],
                color='r', meanline=True)
    axes[0][1].set_title(f'Box plot for "{column}"',
                      size=15)
    
    sns.kdeplot(np.log(data[column] + 1), ax=axes[1][0],
                color='violet', shade=True)
    axes[1][0].set_title(f'Log-distribution for "{column}"',
                      size=15)
    
    sns.boxplot(np.log(data[column] + 1), ax=axes[1][1],
                color='violet', meanline=True)
    axes[1][1].set_title(f'Box plot for logarithmic "{column}"',
                      size=15)

    plt.subplots_adjust(wspace=0.1, hspace=0.3);
    
    
def bar_result(column):
    """
    Plots pie chart and bar chart
    for given categorical feature
    """
    fig, axes = plt.subplots(figsize=(12, 6))
    
    sns.barplot(x=result[column].astype(str).value_counts().index[:5],
                y=result[column].value_counts(normalize=True).values[:5],
                palette='magma')
    plt.title(f'Top categories for "{column}"',
              size=15)
    plt.ylabel('Proportion')
    
    plt.xticks(rotation=90)
    plt.subplots_adjust(wspace=0.2);
```

```python
cat_cols=['channel', 'format', 'cource_tag', 'cource_line', 'video_name', 
          'video_month', 'day_of_week', 'video_category', 'tags', 'description', 'promocode']
```

```python
for feature in cat_cols:
        data_for_model[feature] = data_for_model[feature].astype('string')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for column in cat_cols:
    data_for_model[column] = label_encoder.fit_transform(data_for_model[column])
```

```python
y = data_for_model['efficiency'].values 
X = data_for_model.drop(['efficiency'], axis=1)


from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
RANDOM_SEED=42
VAL_SIZE   = 0.5

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=VAL_SIZE, random_state=RANDOM_SEED)


clf = LogisticRegression(random_state=RANDOM_SEED)
clf.fit(X_train, y_train)
y_pred_prob = clf.predict_proba(X_test)[:,1]
y_pred = clf.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("–ú–æ–¥–µ–ª—å:",'Accuracy(–¥–æ–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤): %.4f' % accuracy_score(y_test, y_pred),
      'Precision(–¥–æ–ª—è –æ–±—ä–µ–∫—Ç–æ–≤ –¢–†): %.4f' % precision_score(y_test,y_pred),
      'Recall(–¥–æ–ª—è P –∏–∑ –≤—Å–µ—Ö –Ω–∞–π–¥–µ–Ω–Ω—ã—Ö –†): %.4f' % recall_score(y_test,y_pred),
      'F1(—Å—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ precision –∏ recall): %.4f' % f1_score(y_test,y_pred))


confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'])
print(confusion_matrix)


from sklearn.metrics import auc, roc_auc_score, roc_curve
fpr, tpr, threshold = roc_curve(y_test, y_pred_prob)
roc_auc = roc_auc_score(y_test, y_pred_prob)

plt.figure()
plt.plot([0, 1], label='Baseline', linestyle='--')
plt.plot(fpr, tpr, label = 'Regression')
plt.title('Logistic Regression ROC AUC = %0.3f' % roc_auc)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc = 'lower right')
plt.show()
```
